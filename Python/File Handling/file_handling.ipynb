```python
# Task 1: Create and write to students.txt

with open("students.txt", "w") as file:
    file.write("Name, Age, Grade\n")
    file.write("John, 20, A\n")
    file.write("Alice, 19, B\n")
    file.write("Mark, 21, A\n")
    file.write("Sophie, 22, C\n")

print("File created and data written successfully.")
```

```python
# Task 2: Read and display file content

with open("students.txt", "r") as file:
    content = file.read()

print("Full File Content:\n")
print(content)

print("Line by Line:\n")
lines = content.split("\n")
for line in lines:
    if line:  # Avoid printing empty lines
        print(line)
```

```python
# Task 3: Append new data

with open("students.txt", "a") as file:
    file.write("Emma, 20, B\n")
    file.write("Liam, 23, A\n")

print("New data appended successfully.")
```

```python
# Task 4: Analyze the file

total_students = 0
grade_a_count = 0

with open("students.txt", "r") as file:
    lines = file.readlines()

for line in lines[1:]:
    if line.strip():  
        total_students += 1
        parts = line.strip().split(", ")
        if parts[2] == "A":
            grade_a_count += 1

print("Total number of students:", total_students)
print("Number of students with grade 'A':", grade_a_count)
```

```python
# Task 5: Update Sophie's age to 23

updated_lines = []

with open("students.txt", "r") as file:
    lines = file.readlines()

for line in lines:
    if line.startswith("Sophie"):
        updated_lines.append("Sophie, 23, C\n")
    else:
        updated_lines.append(line)

with open("students.txt", "w") as file:
    file.writelines(updated_lines)

print("Sophie's age updated successfully.")
```

```python
# Task 6: Work with Large Files

with open("numbers.txt", "w") as file:
    for i in range(1, 1001):
        file.write(str(i) + "\n")

print("numbers.txt created successfully.")

total_sum = 0

with open("numbers.txt", "r") as file:
    for line in file:
        total_sum += int(line.strip())

print("Sum of numbers from 1 to 1000:", total_sum)
```

```python
# Task 7: Error Handling

try:
    with open("non_existent_file.txt", "r") as file:
        content = file.read()
        print(content)

except FileNotFoundError:
    print("Error: The file does not exist.")

except PermissionError:
    print("Error: You do not have permission to access this file.")

except Exception as e:
    print("An unexpected error occurred:", e)

finally:
    print("File operation attempt completed.")
```

```python
# Task 8: Word Count

# Step 1: Create paragraph.txt (Run once)
with open("paragraph.txt", "w") as file:
    file.write("Python is powerful and easy to learn. "
               "Python is widely used in data science, web development, and automation.")

print("paragraph.txt created successfully.")

# Step 2: Count total words and specific word occurrence
with open("paragraph.txt", "r") as file:
    content = file.read()

words = content.split()
total_words = len(words)

print("Total number of words:", total_words)

# Count specific word entered by user
search_word = input("Enter a word to count: ").lower()

word_count = words.count(search_word)

print(f"The word '{search_word}' appears {word_count} times.")
```

```python
# Task 9: Copy File Content

try:
    with open("students.txt", "r") as source:
        with open("students_backup.txt", "w") as destination:
            for line in source:
                destination.write(line)

    print("File copied successfully to students_backup.txt")

except FileNotFoundError:
    print("Error: students.txt does not exist.")

except Exception as e:
    print("An error occurred:", e)
```

```python
# Task 10: Delete a File

import os

file_name = "students_backup.txt"

if os.path.exists(file_name):
    confirm = input(f"Are you sure you want to delete '{file_name}'? (yes/no): ").lower()

    if confirm == "yes":
        os.remove(file_name)
        print("File deleted successfully.")
    else:
        print("File deletion cancelled.")
else:
    print("File does not exist.")
```
