```python

def calculate_salary (base_salary, bonus_percent=10, deductions_percent=5):
 bonus = base_salary * (bonus_percent / 100)
 deductions = base_salary * (deductions_percent / 100)
 total_salary = base_salary + bonus - deductions
 return total_salary

calculate_salary(50000,10,5)
```

```python
def apply_discount(price, discount_function):
    return discount_function(price)


flat_discount = lambda price: price - 50
percentage_discount = lambda price: price * 0.8  

print(apply_discount(500, flat_discount))
print(apply_discount(500, percentage_discount))
```

```python
def fibonacci(n):
    if n <= 1:
        return n
    return fibonacci(n - 1) + fibonacci(n - 2)


print(fibonacci(10))
```

```python
products = [
    {"name": "Laptop", "price": 1200},
    {"name": "Phone", "price": 800},
    {"name": "Tablet", "price": 600},
    {"name": "Monitor", "price": 300}
]

# Apply discount
discounted_prices = list(map(
    lambda p: {"name": p["name"], "price": p["price"] * 0.9},
    products
))

print(discounted_prices)

# Filter products0
expensive_products = list(filter(
    lambda p: p["price"] > 500,
    products
))

print(expensive_products)
```

```python
import time

def execution_time(func):
    def wrapper(*args, **kwargs):
        start = time.time()
        result = func(*args, **kwargs)
        end = time.time()
        print(f'execution time :- {end - start:.6f} seconds')
        return result
    return wrapper

@execution_time
def factorial(n):
    result = 1
    for i in range(1, n + 1):
        result *= i
    return result

factorial(10)
```

```python
from functools import reduce

def custom_aggregate(numbers, func):
    return reduce(func, numbers)

sum_result = custom_aggregate([1, 2, 3, 4], lambda x, y: x + y)
print("Sum:", sum_result)

product_result = custom_aggregate([1, 2, 3, 4], lambda x, y: x * y)
print("Product:", product_result)
```

```python
from functools import reduce

def dynamic_function(numbers, operation):
    operations = {
        "add": lambda x, y: x + y,
        "subtract": lambda x, y: x - y,
        "multiply": lambda x, y: x * y,
        "divide": lambda x, y: x / y if y != 0 else float('inf')
    }
    
    if operation not in operations:
        raise ValueError("Invalid operation")
    
    return reduce(operations[operation], numbers)


print(dynamic_function([10, 5, 2], "add"))
print(dynamic_function([10, 5, 2], "subtract"))
print(dynamic_function([10, 5, 2], "multiply"))
print(dynamic_function([100, 5, 2], "divide"))
```
